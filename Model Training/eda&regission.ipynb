```python
data.drop(['region_1'],axis=1,inplace=True)
data.drop(['region_2'],axis=1,inplace=True)
data.head()
```

```python
df2 = data
```

```python
categorical_columns = df2.select_dtypes(include=['object']).columns
numeric_columns = df2.select_dtypes(include=['float64','int64']).columns
for col in categorical_columns:
        df2[col] = df2[col].astype('category')
df2[categorical_columns] = df2[categorical_columns].apply(lambda x: x.cat.codes)
```

```python
missing_col = df2.isnull().sum(axis=0).reset_index()
missing_col.columns = ['column_name', 'missing_count']
missing_col
```

```python
(X,Y)=(df2.drop('price',axis=1),df2['price'])
(train_x,test_x,train_y,test_y)=train_test_split(X,Y,random_state=50)
```

```python
def mean_absolute_percentage_error(y_true, y_pred): 
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
```

LinearRegression

```python
linear = LinearRegression()
linear.fit(train_x,train_y)
```

```python
pred_y = linear.predict(test_x)
print("Score: "+str(linear.score(test_x,test_y)))
print("RMS: "+str(sqrt(metrics.mean_squared_error(test_y,pred_y))))
print("MAPE: "+str(mean_absolute_percentage_error(test_y,pred_y)))
print("R2: "+str(metrics.r2_score(test_y,pred_y)))
print("MAE: "+str(metrics.mean_absolute_error(test_y,pred_y)))
```

LASSO

```python
lasso= linear_model.Lasso()
lasso.fit(train_x, train_y)
```

```python
pred_y = lasso.predict(test_x)
print("Score for train: "+str(lasso.score(train_x,train_y)))
print("Score: "+str(lasso.score(test_x,test_y)))
print("RMS: "+str(sqrt(metrics.mean_squared_error(test_y,pred_y))))
print("MAPE: "+str(mean_absolute_percentage_error(test_y,pred_y)))
print("R2: "+str(metrics.r2_score(test_y,pred_y)))
print("MAE: "+str(metrics.mean_absolute_error(test_y,pred_y)))
```

mlp

```python
transformer = Normalizer().fit(train_x)
mlp_train_x = pd.DataFrame(transformer.transform(train_x))
mlp_train_x.columns = train_x.columns

mlp_test_x = Normalizer().fit(test_x)
mlp_test_x = pd.DataFrame(transformer.transform(test_x))
mlp_test_x.columns = test_x.columns
```

```python
mlp = MLPRegressor(hidden_layer_sizes=(200,10),learning_rate='adaptive',solver='adam',random_state=5,learning_rate_init=0.1,max_iter=400)
mlp.fit(train_x,train_y)
```

```python
pred_y = mlp.predict(mlp_test_x)
print("Score for train: "+str(mlp.score(mlp_train_x,train_y)))
print("Score: "+str(mlp.score(mlp_test_x,test_y)))
print("RMS: "+str(sqrt(metrics.mean_squared_error(test_y,pred_y))))
print("MAPE: "+str(mean_absolute_percentage_error(test_y,pred_y)))
print("R2: "+str(metrics.r2_score(test_y,pred_y)))
print("MAE: "+str(metrics.mean_absolute_error(test_y,pred_y)))
```
